### Cross correlations above show Human Capital & Research & Business Sophistication are correlated with knowledge & Technology outputs

#### Some detailed analysis

```python
dfs_header.plot.scatter(x='Business sophistication', y='Knowledge and technology outputs')
```

```python
#Want to know how the coutnries are clustered 
clusterer = sklearn.cluster.KMeans(n_clusters=5)
```

```python
cluster_suggestions = clusterer.fit_predict(dfs_header[['Business sophistication', 'Knowledge and technology outputs']])
cluster_suggestions
```

```python
clusterer.cluster_centers_
```

```python
X = dfs_header['Business sophistication']
y = dfs_header['Knowledge and technology outputs']
```

```python
#What are the clustered countries?
colouring = np.array(['yellow', 'blue', 'red', 'purple', 'black'])
color_by_cluster = colouring[cluster_suggestions]
# clour_by_cluster
clusters_df = pd.DataFrame({'Labels': X.index, 'Colours': color_by_cluster})
clusters_df.sort_values('Colours', inplace=True, ascending=False)

clusters_df
```

```python
matplotlib.pyplot.scatter(X,y, color = color_by_cluster)
matplotlib.pyplot.xlabel('Business sophistication')
matplotlib.pyplot.ylabel('Knowledge and technology outputs')
matplotlib.pyplot.title('Clusters', size=18)
```
