```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
import sklearn.cluster
import matplotlib
```

# Importing two CSV files I created in Global_Innovation_CSV.  

```python
#Concatenated countries from Global Innovation Index site https://www.globalinnovationindex.org/analysis-economy
Global_innovation = pd.read_csv('Global_II.CSV', index_col=0)
Global_innovation.fillna(Global_innovation.mean(), inplace=True)
Global_innovation.sort_index(inplace=True)
Global_innovation
```

```python
## This cell limits my dataframe to Institutions that have greater than 90%
Global_innovation[Global_innovation.Institutions > 90]
```

### Global_innovation.csv will now be split into two dataframes

```python
##. Only want the indicators for innovation ouput

Innovation_output = Global_innovation[Global_innovation.columns[72:109]] 
```

```python
##. Only want the indicators for innovation input

Innovation_input = Global_innovation[Global_innovation.columns[0:72]] 
```

####    Innovation_index includes the average sum of all measures that indicate technical innovation input, output and their ratio (output/input) of each country.

```python
Innovation_index = pd.read_csv('Global_Index.csv', index_col=0)
Innovation_index.sort_index(inplace=True)
Innovation_index.fillna(Innovation_index.mean(), inplace=True)
```

# Analysing Data :

          Individual innovation data vs Innovation index* 
                        a). Innovation_index[Innovation efficiency ratio] vs Global_innovation
                        b). Innovation_index[Innovation input subindex] vs Innovation_output
                        c). Innovation_index[Innovation output subindex] vs Innovation_input

###     a). Innovation efficiency ratio

```python
import sklearn.linear_model
from sklearn import model_selection
from sklearn.linear_model import ElasticNet
prediction_columns = Global_innovation.columns
X = Global_innovation[prediction_columns]
y = Innovation_index['Innovation Efficiency Ratio']
```

```python
elastic = sklearn.linear_model.ElasticNetCV()
elastic.fit(X,y)
elastic2 = zip(elastic.coef_, prediction_columns)
elastic2
```

```python
#returns coefficient of determination R^2 of the predicition
elastic.score(X, y)
```

```python
sorted(elastic2)
```

```python
import sklearn.tree
import sklearn.grid_search
```

```python
brute_force = sklearn.grid_search.GridSearchCV(
    sklearn.tree.DecisionTreeRegressor(),
    param_grid = {
        'max_depth': [1,2,3,4,5],
        'criterion': ['mse']
    }
) 
```

```python
brute_force.fit(X, y)
```

```python
# This doesn't give a consistent parameter
brute_force.best_params_
```

```python
dtc = sklearn.tree.DecisionTreeRegressor(max_depth=3)
dtc.fit(X, y)
```

```python
try:
 from StringIO import StringIO
except ImportError:
 from io import StringIO
import sklearn.tree
import IPython.display
import pydot
File_obj = StringIO()
sklearn.tree.export_graphviz(dtc, out_file=File_obj,feature_names=X.columns)
Graph = pydot.graph_from_dot_data(File_obj.getvalue())
IPython.display.Image(Graph.create_png())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
from sklearn.cross_validation import cross_val_score



model = RandomForestRegressor(100, oob_score=True, n_jobs=-1, random_state=42)
model.fit(X, y)

features = X.columns
feature_importances = model.feature_importances_
features_df = pd.DataFrame({'Features': features, 'Importance Score': feature_importances})
features_df.sort_values('Importance Score', inplace=True, ascending=False)

features_df.head()
```

```python
from sklearn.cross_validation import cross_val_score
X = Global_innovation
y = Innovation_index['Innovation Efficiency Ratio']

scores = cross_val_score(model, X, y, scoring='mean_squared_error')
print('CV MSE {}, Average MSE {}'.format(scores, scores.mean()))

for n_trees in range(1, 100, 10):
    model = RandomForestRegressor(n_estimators = n_trees)
    scores = cross_val_score(model, X, y, scoring='mean_squared_error')
    print('n trees: {}, CV MSE {}, Average MSE {}'.format(n_trees, scores, scores.mean()))
    
    model.fit(X, y)

# Get columns and their scores
features = X.columns
feature_importances = model.feature_importances_
features_df = pd.DataFrame({'Features': features, 'Importance Score': feature_importances})
features_df.sort_values('Importance Score', inplace=True, ascending=False)

features_df.head()
```
